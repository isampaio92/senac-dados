```python
import pandas as pd
import numpy as np
```

```python
def calcularDistancia(df, coluna):
    array = np.array(df[coluna])
    media = np.mean(array)
    mediana = np.median(array)
    distancia = (media - mediana) / mediana
    return distancia
```

```python
def agruparMedia(df, qualitativo, quantitativo, menorParaMaior):
    return df.groupby(qualitativo)[quantitativo].mean().reset_index().sort_values(by=quantitativo, ascending=menorParaMaior)
```

```python
def calcularQuartis(df, coluna):
    array = np.array(df[coluna])
    q1 = np.percentile(array, 25)
    q3 = np.percentile(array, 75)
    return [q1, q3]
```
