```python
import pandas as pd
import numpy as np
```

```python
dados = 'CARAMELO2.csv'
df = pd.read_csv(dados, sep=';')
```

```python
df.head()
```

```python
df['Valor de Venda'] = df['Valor de Venda'].str.replace(',', '.').astype(float)
df.head()
```

```python
df_venda_loja = df.groupby('Loja')['Valor de Venda'].sum().reset_index().sort_values(by='Valor de Venda', ascending=False)
df_venda_loja
```

```python
array_venda_loja = np.array(df_venda_loja['Valor de Venda'])
q1_venda_loja = np.percentile(array_venda_loja, 25)
q2_venda_loja = np.percentile(array_venda_loja, 50)
q3_venda_loja = np.percentile(array_venda_loja, 75)
media_venda_loja = np.mean(array_venda_loja)
distancia_venda_loja = (media_venda_loja - q2_venda_loja) / q2_venda_loja
iqr_venda_loja = q3_venda_loja - q1_venda_loja
lim_sup_venda_loja = q3_venda_loja + (1.5 * iqr_venda_loja)
lim_inf_venda_loja = q1_venda_loja - (1.5 * iqr_venda_loja)

print(f'Q1: {q1_venda_loja}')
print(f'Q2: {q2_venda_loja}')
print(f'Q3: {q3_venda_loja}')
print(f'Média: {media_venda_loja}')
print(f'Mediana: {q2_venda_loja}')
print(f'Distância: {distancia_venda_loja}')
print(f'IQR: {iqr_venda_loja}')
print(f'Limite Superior: {lim_sup_venda_loja}')
print(f'Limite Inferior: {lim_inf_venda_loja}')
```

```python
outliers_venda_loja = df_venda_loja.loc[(df_venda_loja['Valor de Venda'] > lim_sup_venda_loja) | (df_venda_loja['Valor de Venda'] < lim_inf_venda_loja)]
outliers_venda_loja
```
