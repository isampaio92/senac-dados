```python
dados_numpy = np.array([12, 15, 17, 20, 22, 25, 28, 30, 35, 40])

q1 = np.percentile(dados_numpy, 25)
q2 = np.percentile(dados_numpy, 50)
q3 = np.percentile(dados_numpy, 75)

print(f'Primeiro quartil (Q1): {q1}')
print(f'Segundo quartil (Q2): {q2}')
print(f'Terceiro quartil (Q3): {q3}')
```

```python
array_furto_celular = np.array(df_furto_celular_cisp['furto_celular'])

q1_furto_celular = np.percentile(array_furto_celular, 25)
q2_furto_celular = np.percentile(array_furto_celular, 50)
q3_furto_celular = np.percentile(array_furto_celular, 75)
media_furto_celular = np.mean(array_furto_celular)
mediana_furto_celular = np.median(array_furto_celular)
distancia_furto_celular = (media_furto_celular - mediana_furto_celular) / mediana_furto_celular

print(f'Primeiro quartil (Q1): {q1_furto_celular}')
print(f'Segundo quartil (Q2): {q2_furto_celular}') # Mediana
print(f'Terceiro quartil (Q3): {q3_furto_celular}')
print(f'Média: {media_furto_celular}')
print(f'Mediana: {mediana_furto_celular}')
print(f'Distância entre média e mediana: {distancia_furto_celular}')
```

```python
df_menor_q1_furto_celulares = df_furto_celular_cisp.loc[df_furto_celular_cisp['furto_celular'] < q1_furto_celular]

df_menor_q1_furto_celulares = df_menor_q1_furto_celulares.sort_values(by='furto_celular')
#df_menor_q1_furto_celulares = df_menor_q1_furto_celulares.sort_values(by='furto_celular', ascending= False)

df_menor_q1_furto_celulares
```

```python
df_capital_2020 = df.loc[(df['ano'] == 2020) & (df['regiao'] == 'Capital')]
df_apreenssao_menores_capital_2020 = df_capital_2020.groupby('cisp')['aaapai'].sum().reset_index()

array_apreensao_menores_capital_2020 = np.array(df_apreenssao_menores_capital_2020['aaapai'])

q1_apreensao_menores_capital_2020 = np.percentile(array_apreensao_menores_capital_2020, 25)
q2_apreensao_menores_capital_2020 = np.percentile(array_apreensao_menores_capital_2020, 50)
q3_apreensao_menores_capital_2020 = np.percentile(array_apreensao_menores_capital_2020, 75)
```

```python
df_maior_q3_apreensao_menores_capital_2020 = df_apreenssao_menores_capital_2020.loc[df_apreenssao_menores_capital_2020['aaapai'] > q3_apreensao_menores_capital_2020]
df_maior_q3_apreensao_menores_capital_2020 = df_maior_q3_apreensao_menores_capital_2020.sort_values(by='aaapai', ascending=False)

df_maior_q3_apreensao_menores_capital_2020
```

```python
regiao_metropolitana = ['Belford Roxo', 'Cachoeiras de Macacu', 'Duque de Caxias', 'Guapimirim', 'Itaboraí', 'Itaguaí', 'Japeri', 'Magé', 'Maricá', 'Mesquita', 'Nilópolis', 'Niterói', 'Nova Iguaçu', 'Paracambi', 'Petrópolis', 'Queimados', 'Rio Bonito', 'Rio de Janeiro', 'São Gonçalo', 'São João de Meriti', 'Seropédica', 'Tanguá']
df_metropolitana_2024 = df.loc[(df['ano'] == 2024) & (df['munic'].isin(regiao_metropolitana))]
df_furto_cel_rm_24 = df_metropolitana_2024.groupby('cisp')['furto_celular'].sum().reset_index()

array_furtos_celular_metr_2024 = np.array(df_furto_cel_rm_24['furto_celular'])
media_furto_celular_metr_2024 = np.mean(array_furtos_celular_metr_2024)
mediana_furto_celular_metr_2024 = np.median(array_furtos_celular_metr_2024)
distancia_furto_celular_metr_2024 = (media_furto_celular_metr_2024 - mediana_furto_celular_metr_2024) / mediana_furto_celular_metr_2024

df_furto_cel_rm_24
```

```python
print(f'Média de furtos de celular na região metropolitana: {media_furto_celular_metr_2024}')
print(f'Mediana de furtos de celular na região metropolitana: {mediana_furto_celular_metr_2024}')
print(f'Distância entre média e mediana: {distancia_furto_celular_metr_2024}')
print(f'Distância em porcentagem: {round(distancia_furto_celular_metr_2024, 2) * 100}%')
```
