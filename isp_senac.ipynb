```python
dados_numpy = np.array([12, 15, 17, 20, 22, 25, 28, 30, 35, 40])

q1 = np.percentile(dados_numpy, 25)
q2 = np.percentile(dados_numpy, 50)
q3 = np.percentile(dados_numpy, 75)

print(f'Primeiro quartil (Q1): {q1}')
print(f'Segundo quartil (Q2): {q2}')
print(f'Terceiro quartil (Q3): {q3}')
```

```python
array_furto_celular = np.array(df_furto_celular_cisp['furto_celular'])

q1_furto_celular = np.percentile(array_furto_celular, 25)
q2_furto_celular = np.percentile(array_furto_celular, 50)
q3_furto_celular = np.percentile(array_furto_celular, 75)
media_furto_celular = np.mean(array_furto_celular)
mediana_furto_celular = np.median(array_furto_celular)
distancia_furto_celular = (media_furto_celular - mediana_furto_celular) / mediana_furto_celular
iqr_furto_celular = q3_furto_celular - q1_furto_celular
lim_sup_furto_celular = q3_furto_celular + (1.5 * iqr_furto_celular)
lim_inf_furto_celular = q1_furto_celular - (1.5 * iqr_furto_celular)

print(f'Primeiro quartil (Q1): {q1_furto_celular}')
print(f'Segundo quartil (Q2): {q2_furto_celular}') # Mediana
print(f'Terceiro quartil (Q3): {q3_furto_celular}')
print(f'Média: {media_furto_celular}')
print(f'Mediana: {mediana_furto_celular}')
print(f'Distância entre média e mediana: {distancia_furto_celular}')
print(f'Distância em porcentagem: {distancia_furto_celular * 100}%')
print(f'Intervalo entre quartis (IQR): {iqr_furto_celular}')
print(f'Limite superior: {lim_sup_furto_celular}')
print(f'Limite inferior: {lim_inf_furto_celular}')
```

```python
df_furto_celular_cisp_outliers = df_furto_celular_cisp.loc[df_furto_celular_cisp['furto_celular'] >= lim_sup_furto_celular]
df_furto_celular_cisp_outliers = df_furto_celular_cisp_outliers.sort_values(by='furto_celular', ascending=False)
df_furto_celular_cisp_outliers
```

```python
df_menor_q1_furto_celulares = df_furto_celular_cisp.loc[df_furto_celular_cisp['furto_celular'] < q1_furto_celular]

df_menor_q1_furto_celulares = df_menor_q1_furto_celulares.sort_values(by='furto_celular')
#df_menor_q1_furto_celulares = df_menor_q1_furto_celulares.sort_values(by='furto_celular', ascending= False)

df_menor_q1_furto_celulares
```
