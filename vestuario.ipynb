```python
import pandas as pd
```

```python
dados = 'vendas_vestuario_revisao.csv'
df = pd.read_csv(dados)

df.head()
```

```python
df['valor_vendido'] = df['quantidade'] * df['valor_unitario']
df
```

```python
df_vendido_vendedor = df.groupby('vendedor')['valor_vendido'].sum().reset_index().sort_values(by='valor_vendido')
df_vendido_vendedor
```
